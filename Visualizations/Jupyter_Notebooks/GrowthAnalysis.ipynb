```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

data = pd.read_csv("MergedFips.csv")
```

```python
LA = data.loc[data['county_name'] == "los angeles, ca"]
# Marion = data.loc[data['county_name'] == "marion, or"]

print(LA['average_listing_price_mm'].mean())
# print(Marion['average_listing_price_mm'].mean())

plt.figure(figsize=(10, 4))  # Setting the figure size
sns.scatterplot(data=LA, x='Date', y='average_listing_price_mm', label='LA', color='blue')
sns.lineplot(data=LA, x='Date', y='average_listing_price_mm', label='LA', color='blue')
plt.axhline(y=0,linewidth=1, color='black')
plt.xlabel('Date')
plt.ylabel('Monthly Percentage Growth')
plt.title('Los Angeles, CA - Monthly Percentage Growth from 2020 - 20224')
plt.xticks(rotation=90)
plt.gca().invert_xaxis()
plt.show()
```

```python
NJ = data.loc[data['State'] == "NJ" ]
NJG = NJ.groupby("Date")["average_listing_price_mm"].mean()
NJG = pd.DataFrame({'Date':NJG.index, 'average_listing_price_mm':NJG.values})

sns.set(style="whitegrid")  # Setting the style to whitegrid for a clean background
 
plt.figure(figsize=(10, 4))  # Setting the figure size
sns.lineplot(data=NJG, x='Date', y='average_listing_price_mm', label='New Jersey', color='green')
sns.scatterplot(data=NJG, x='Date', y='average_listing_price_mm', label='New Jersey', color='green')
plt.xlabel('Date')
plt.ylabel('Monthly Percentage Growth')
plt.title('State - Monthly Percentage Growth from 2020 - 20224')
plt.xticks(rotation=90)
plt.show()


sns.set(style="whitegrid")  # Setting the style to whitegrid for a clean background
plt.figure(figsize=(14, 8))  # Setting the figure size

sns.lineplot(data=NJ, x="Date", y="average_listing_price_mm", hue="county_name")

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.xlabel('Date')
plt.ylabel('Monthly Percentage Growth')
plt.title('NJ Counties - Zillow Value from 2020 - 20224')
plt.xticks(rotation=90)
plt.gca().invert_xaxis()
plt.show()
```
